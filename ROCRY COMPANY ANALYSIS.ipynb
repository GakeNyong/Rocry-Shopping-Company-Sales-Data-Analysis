```python
import pandas as pd
df = pd.read_excel('RSC.xlsx')
df
```

```python
# Question 1 - Delete the empty columns (Status and unnamed1)
del df['Status']
del df['unnamed1']
df
```

```python
# Question 2 - On the Gender Column, Replace the M and F to Male and Female Respectively
df['Gender'] = df['Gender'].str.replace('M', "Male")
df['Gender'] = df['Gender'].str.replace('F', "Female")
df
```

```python
# Question 3 -	On the Marital Status Column, replace 0 and 1 to Single and Married respectively
df['Marital_Status'] = df['Marital_Status'].astype(str)
df['Marital_Status'] = df['Marital_Status'].str.replace('0', 'Single')
df['Marital_Status'] = df['Marital_Status'].str.replace('1', 'Married')
df
```

### Some Orders were Unsuccessful i.e., incomplete. They are the ones with empty records in the Amount Columns.

```python
# Question 4 - Replace such empty cells with 0
df['Amount'] = df['Amount'].fillna(0)
df.head(10)
```

### Using appropriate skill get the following informations

```python
# Question 5 - Total number of Customers; Not Distinct; df['Cust_name'].count()
df['Cust_name'].nunique()
```

```python
# Sum of Orders and GrandTotal
df.pivot_table(index = "Occupation",
               values = "Orders",
               aggfunc = "sum",
              margins = True,
              margins_name = 'Grand_Total')
```

```python
# Question 6 - What Occupation placed the highest numbers of Orders and by what Sum of Orders
Q6 = df.pivot_table(index = "Occupation",
               values = "Orders",
               aggfunc = "sum")
Q6.sort_values(by=['Orders'], ascending = False)
```

```python
# Question 7 - The Top 10 Product_Category by Sum of Amount and the corresponding percentage of total
Q7 = df.pivot_table(index = "Product_Category",
               values = "Amount",
               aggfunc = "sum")

Q7 = Q7.sort_values(by=['Amount'], ascending=False).head(10)
Q7["%"]=100*Q7['Amount']/Q7['Amount'].sum()
Q7.round(0)
```

```python
# Question 8 - What Age Group of Male and Female respectively placed the highest number of Orders
Q8 = df.pivot_table(index= ['Gender', 'Age Group'],
                    values = "Orders",
                    aggfunc= "sum")
Q8.sort_values(['Gender', 'Orders'],ascending = False).groupby('Gender').head()
```

```python
#Sort from Lowest to Highest
Q8 = df.pivot_table(index= ['Gender', 'Age Group'],
                    values = "Orders",
                    aggfunc= "sum")
Q8.sort_values(['Gender', 'Orders']).groupby('Gender').head()
```
